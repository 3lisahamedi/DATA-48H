```python
import pandas as pd

pd.options.display.max_rows = 9999  # Afficher plus de lignes

# Lire le fichier CSV avec pr√©caution
try:
    df = pd.read_csv("tweets.csv", sep=";", encoding="utf-8") 
    # ici le sep ; permet une bonne lecture du fichier, j'ai remarque sa en observant les fichier csv,
    # il y avait une erreur de lecture
    print(df)
        
    # Gestion d'erreu ===
except FileNotFoundError:
    print("Erreur : Le fichier 'tweets.csv' est introuvable.")
except pd.errors.ParserError:
    print("Erreur : Probl√®me de format dans le fichier CSV.")
except Exception as e:
    print(f"Erreur inattendue : {e}")

    
    # Supprimer les colonnes souhait√©es (ex: "colonne1" et "colonne2")
df_modifie = df.drop(columns=["id", "screen_name", "name"])  

# Sauvegarder dans un nouveau fichier sans modifier l'original
df_modifie.to_csv("tweets_v2.csv", index=False, encoding="utf-8")

print("Le nouveau fichier 'tweets_v2.csv' a √©t√© cr√©√© avec succ√®s !")
```
